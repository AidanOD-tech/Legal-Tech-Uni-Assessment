      a) In the first round, faculty members are asked to pick their top choice.  The winner is ranked the first-candidate.

      b) The winner of this first election is removed from the second ballot and faculty members are asked to make their selection of their most-preferred candidate among those that remain.  The winner of this round is ranked the second-candidate.

      c) The winner of the second round is removed from the final ballot and faculty members are asked to select which of the two remaining candidates they prefer.  The winner of this round is the third-candidate and the loser of this round is the fourth-candidate.
